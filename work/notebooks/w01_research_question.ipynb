## 3. Quick look at the data (2-3 real numbers)

I analyzed the starter dataset of 30,000 rows. The data confirms the scale of the content decay problem:

1.  **54.21% of all pages** (16,262 out of 30,000) are currently in a declining traffic trend (`trend_direction == "down"`).
2.  These declining pages account for **51.27% of all search impressions** (79.99M out of 156.01M total impressions) over the last 90 days, indicating a massive volume of traffic currently at risk.
3.  The average days since the last update for declining pages is **49.2 days** (compared to 46.1 days overall), confirming a link between content staleness and traffic decay.

```python
import pandas as pd

# Load starter dataset
raw_url = "https://raw.githubusercontent.com/Sizan99/ml-pipeline/main/data/raw/content_refresh_anonymized.csv"
print(f"Loading data from: {raw_url}")
df = pd.read_csv(raw_url)

total_pages = len(df)
declining_pages = len(df[df["trend_direction"].str.lower() == "down"])
pct_declining = (declining_pages / total_pages) * 100

total_impressions = df["impressions_90d"].sum()
declining_impressions = df[df["trend_direction"].str.lower() == "down"]["impressions_90d"].sum()
pct_impressions_at_risk = (declining_impressions / total_impressions) * 100

avg_days_since_update_all = df["days_since_last_update"].mean()
avg_days_since_update_declining = df[df["trend_direction"].str.lower() == "down"]["days_since_last_update"].mean()

print(f"Total Pages: {total_pages:,}")
print(f"Declining Pages: {declining_pages:,} ({pct_declining:.2f}%)")
print(f"Total Impressions: {total_impressions:,}")
print(f"Impressions at Risk: {declining_impressions:,} ({pct_impressions_at_risk:.2f}%)")
print(f"Average Days Since Last Update (All): {avg_days_since_update_all:.1f} days")
print(f"Average Days Since Last Update (Declining): {avg_days_since_update_declining:.1f} days")
```
